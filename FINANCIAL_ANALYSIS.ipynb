```python
# Filter good loans
good_loans = df[df['loan_status'].isin(['Fully Paid', 'Current'])]

# Total loan applications
total_loan_applications = df['id'].count()

# Good loan applications
good_loan_applications = good_loans['id'].count()

# Good loan funded amount
good_loan_funded_amount = good_loans['loan_amount'].sum()

# Good loan amount received
good_loan_received = good_loans['total_payment'].sum()

# Convert to millions
good_loan_funded_amount_millions = good_loan_funded_amount / 1_000_000
good_loan_received_millions = good_loan_received / 1_000_000

# Percentage of good loans
good_loan_percentage = (good_loan_applications / total_loan_applications) * 100

# Print results
print("Good Loan Applications:", good_loan_applications)
print(f"Good Loan Funded Amount (in Millions): ${good_loan_funded_amount_millions:.2f}M")
print(f"Good Loan Total Received (in Millions): ${good_loan_received_millions:.2f}M")
print(f"Percentage of Good Loan Applications: {good_loan_percentage:.2f}%")
```

```python
# Filter bad loans
bad_loans = df[df['loan_status'] == 'Charged Off']

# Total loan applications
total_loan_applications = df['id'].count()

# Bad loan applications
bad_loan_applications = bad_loans['id'].count()

# Bad loan funded amount
bad_loan_funded_amount = bad_loans['loan_amount'].sum()

# Bad loan amount received
bad_loan_received = bad_loans['total_payment'].sum()

# Convert to millions
bad_loan_funded_amount_millions = bad_loan_funded_amount / 1_000_000
bad_loan_received_millions = bad_loan_received / 1_000_000

# Percentage of bad loans
bad_loan_percentage = (bad_loan_applications / total_loan_applications) * 100

# Print results
print("Bad Loan Applications:", bad_loan_applications)
print(f"Bad Loan Funded Amount (in Millions): ${bad_loan_funded_amount_millions:.2f}M")
print(f"Bad Loan Total Received (in Millions): ${bad_loan_received_millions:.2f}M")
print(f"Percentage of Bad Loan Applications: {bad_loan_percentage:.2f}%")
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

sns.set_style("whitegrid")

# Ensure date column is datetime
df['issue_date'] = pd.to_datetime(df['issue_date'])
```

```python
monthly_trend = (
    df.groupby(df['issue_date'].dt.to_period('M'))
      .agg(
          Loan_Applications=('id', 'count'),
          Funded_Amount=('loan_amount', 'sum')
      )
      .reset_index()
)

monthly_trend['issue_date'] = monthly_trend['issue_date'].astype(str)

plt.figure(figsize=(12,6))
plt.plot(monthly_trend['issue_date'], monthly_trend['Loan_Applications'], label='Loan Applications')
plt.plot(monthly_trend['issue_date'], monthly_trend['Funded_Amount']/1_000_000, label='Funded Amount (M)')
plt.xticks(rotation=45)
plt.title("Monthly Loan Trends by Issue Date")
plt.xlabel("Month")
plt.ylabel("Count / Amount")
plt.legend()
plt.show()
```

```python
import plotly.express as px

state_analysis = (
    df.groupby('address_state')
      .agg(
          Loan_Count=('id', 'count'),
          Total_Funded=('loan_amount', 'sum')
      )
      .reset_index()
      .sort_values(by='Loan_Count', ascending=False)
)

state_analysis['Total_Funded_M'] = state_analysis['Total_Funded'] / 1_000_000

fig = px.bar(
    state_analysis,
    x='Loan_Count',
    y='address_state',
    orientation='h',
    title='Regional Loan Distribution by State',
    hover_data={
        'address_state': True,
        'Loan_Count': True,
        'Total_Funded_M': ':.2f'
    },
    labels={
        'address_state': 'State',
        'Loan_Count': 'Loan Applications',
        'Total_Funded_M': 'Total Funded Amount (M)'
    }
)

# Highest values at top
fig.update_layout(
    yaxis={'categoryorder': 'total ascending'},
    height=800   # important when many states
)

fig.show()
```

```python
import plotly.express as px

# Aggregate data
term_funding = (
    df.groupby('term')['loan_amount']
      .sum()
      .reset_index()
)

# Convert to millions
term_funding['loan_amount_m'] = term_funding['loan_amount'] / 1_000_000

# Create donut chart
fig = px.pie(
    term_funding,
    names='term',
    values='loan_amount_m',
    hole=0.4,
    title='Loan Term Analysis by Total Funded Amount (in Millions)',
    hover_data={
        'loan_amount_m': ':.2f'
    },
    labels={
        'term': 'Loan Term',
        'loan_amount_m': 'Total Funded Amount (M)'
    }
)

# Improve tooltip display
fig.update_traces(
    textinfo='percent+label',
    hovertemplate=
    "<b>Loan Term:</b> %{label}<br>" +
    "<b>Total Funded:</b> $%{value:.2f}M<br>" +
    "<b>Share:</b> %{percent}<extra></extra>"
)

fig.show()
```

```python
import plotly.express as px

# Aggregate funded amount by employment length
emp_funding = (
    df.groupby('emp_length')['loan_amount']
      .sum()
      .reset_index()
      .sort_values(by='loan_amount', ascending=False)
)

# Convert to thousands
emp_funding['Funded_Amount_K'] = emp_funding['loan_amount'] / 1_000

# Plot
fig = px.bar(
    emp_funding,
    x='Funded_Amount_K',
    y='emp_length',
    orientation='h',
    title='Total Funded Amount by Employment Length',
    hover_data={
        'emp_length': True,
        'Funded_Amount_K': ':.1f'
    },
    labels={
        'emp_length': 'Employment Length',
        'Funded_Amount_K': 'Funded Amount (Thousands)'
    }
)

# Ensure highest value at top
fig.update_layout(
    yaxis={'categoryorder': 'total ascending'},
    height=600
)

# Custom tooltip text
fig.update_traces(
    hovertemplate=
    "<b>Employment Length:</b> %{y}<br>" +
    "<b>Total Funded:</b> $%{x:.1f}K<extra></extra>"
)

fig.show()
```

```python
import matplotlib.pyplot as plt
import mplcursors

# Aggregate data (in millions)
purpose_funding_millions = (
    df.groupby('purpose')['loan_amount']
      .sum()
      .sort_values() / 1_000_000
)

# Plot
plt.figure(figsize=(10, 6))
bars = plt.barh(
    purpose_funding_millions.index,
    purpose_funding_millions.values
)

# Labels on bars
for bar in bars:
    width = bar.get_width()
    plt.text(
        width + 0.1,
        bar.get_y() + bar.get_height()/2,
        f'{width:.2f}M',
        va='center',
        fontsize=9
    )

# Title & labels
plt.title('Total Funded Amount by Loan Purpose (₹ Millions)', fontsize=14)
plt.xlabel('Funded Amount (₹ Millions)')
plt.ylabel('Loan Purpose')

plt.grid(axis='x', linestyle='--', alpha=0.6)
plt.tight_layout()


cursor = mplcursors.cursor(bars, hover=True)
cursor.connect(
    "add",
    lambda sel: sel.annotation.set_text(
        f"Purpose: {purpose_funding_millions.index[sel.index]}\n"
        f"Amount: ₹{purpose_funding_millions.values[sel.index]:.2f}M"
    )
)

plt.show()
```

```python
import plotly.express as px

# Aggregate data
home_funding = (
    df.groupby('home_ownership')['loan_amount']
      .sum()
      .reset_index()
)

# Convert to millions
home_funding['loan_amount_millions'] = home_funding['loan_amount'] / 1_000_000

# Create treemap with custom hover
fig = px.treemap(
    home_funding,
    path=['home_ownership'],
    values='loan_amount_millions',
    color='loan_amount_millions',
    color_continuous_scale='Blues',
    title='Total Funded Amount by Home Ownership (₹ Millions)',
    hover_data={
        'loan_amount_millions': ':.2f'
    }
)

# Custom tooltip text
fig.update_traces(
    hovertemplate=
    "<b>Home Ownership:</b> %{label}<br>" +
    "<b>Total Funded:</b> ₹%{value:.2f}M<br>" +
    "<extra></extra>"
)

fig.show()
```
